```python
from PIL import Image 
import numpy as np 
```

```python
# load the image 
im=Image.open("colors.jpeg")
im.show()
```

```python
# convert the image to grayscale 
gray_im = im.convert("L")
gray_im.show()
```

```python
# Test the function Image.split on the  image
r, g, b = im.split()

r.show()
g.show()
b.show()
```

```python
# or numpy_based split
data = np.array(im)

r = data[:,:,0] # red
g = data[:,:,1] # green 
b = data[:,:,2] # blue

new_im_r =Image.fromarray(r)
new_im_g =Image.fromarray(g)
new_im_b =Image.fromarray(b)

new_im_r.show()
new_im_g.show()
new_im_b.show()
```

```python
# Combine the output using merge and display
merged = Image.merge("RGB",(r, g, b))
merged.show()
```

```python
# Replace the green channel with zeros
zero_green = Image.merge("RGB",(r,Image.new("L",g.size,0),b))
zero_green.show()
```
